## I. Puzzle

I have a large, hemispherical piece of bread with a radius of 1 foot. I make a bread bowl by boring out a cylindrical hole with radius r, centered at the top of the hemisphere and extending all the way to the flat bottom crust.

What should the radius of my borehole be to maximize the volume of soup my bread bowl can hold?

# II. Solution

  

Let us call the radius of the borehole $x$. The volume $V$ of the soup is given by the area of the borehole and the height $h$ of the hole:
$$V = \pi x^2 \cdot h$$

The height $h$ is given by the radius of the bread (1) and the radius $x$ (see image below):

$$1 = h^2 + x^2 \Leftrightarrow h = \sqrt{1-x^2}$$

The volume of the soup is maximized when $\mathrm{f}'_V (x) = 0$ 

$$\mathrm{f}'_V (x) = \frac{\mathrm{d}}{\mathrm{d}x}\Big(\pi x^2 \sqrt{1-x^2}\Big)= \frac{\pi x (2-3x^2)}{\sqrt{1-x^2}}$$

With $x>0$ the maximum is at $x=\sqrt{2/3}\approx 0.816$ which gives a volume of 1.21 footÂ³.

# III. Scratch pad

```python
# f_1 
import sympy as sp

x = sp.symbols('x')
f = sp.pi * x**2 * sp.sqrt(1-x**2)
f_1=sp.simplify(sp.diff(f,x))
f_1
```

```python
# maximum
exp = sp.Eq(f_1,0)
sol = sp.solve(exp, x, 0.5) 
sol
```

```python
sol[2][0].round(3)
```
