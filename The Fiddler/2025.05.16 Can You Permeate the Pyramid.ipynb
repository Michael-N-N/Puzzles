## I. Puzzle

Consider the following figure, which shows 16 points arranged in a rhombus shape and connected to their neighbors by edges. How many distinct paths are there from the top point to the bottom along the edges such that:

- You never visit the same point twice.
- You only move downward or sidewaysâ€”never upward

# II. Solution

For each knot in the figure we will calculate the number of possible paths from the uppermost knot to a given knot. The set of possible paths to a given knot can be divided into two subsets depending on the position of the preceding knot in a path. 

The first subset include paths where the last movement is downward. Hence, the preceding knot is on a level above. The size of this subset is the sum of all possible paths to the previous knots.

The second subset consists of paths where the last movement is sideway. As we can move left or right a given knot can be reach from all other knots on the same level. Further as we can't turn around (i.e. visit a knot more than once), the size of the subset is the sum of all first subset of the all knots being on the same level as the given knot.

The number of possible paths from top to bottom is 2,304.
